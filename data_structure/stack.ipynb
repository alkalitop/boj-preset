# Stack

## 개요
후입선출(LIFO, Last-In First-Out)의 특징을 가진 자료구조.
[자세한 정보](https://roi-data.com/entry/%EC%9E%90%EB%A3%8C%EA%B5%AC%EC%A1%B0-4-%EC%8A%A4%ED%83%9DStack%EC%9D%B4%EB%9E%80-%EC%97%B0%EC%82%B0-%EA%B5%AC%ED%98%84%EB%B0%A9%EB%B2%95)

## 구현
Python에서는 일반적으로 list로 구현합니다.

### 스택 생성

```python
stk = []
```

### pop()
스택의 top에 있는 요소를 제거합니다.

```python
stk.pop()
```

### push(item)
스택에 요소를 추가합니다.

```python
stk.append(item)
```

### peek()
스택의 top에 있는 요소를 제거하지 않고 반환합니다.

```python
stk[-1]
```

### size()
스택의 크기를 반환합니다.

```python
len(stk)
```
